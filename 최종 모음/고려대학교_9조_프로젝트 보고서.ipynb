## 연구 목적 및 배경



영국 옥스포드대학교 부설 로이터저널리즘 연구소에서 발간한 ‘디지털뉴스리포트 2019’에 따르면,  
한국의 언론은 2016년에 처음으로 포함이 된 이후로 4년 연속 언론 신뢰도 최하위라는 불명예의 기록을 가지고 있는 것으로 확인이 되었다.    
가장 최근의 기록인 2019년의 경우 언론 신뢰도가 22%로 평균 기록인 42%의 약 절반 수준에 그치는 것으로 확인이 되고 있다. 

이러한 언론 신뢰도가 최악의 수준임을 짐작할 수 있는 또다른 이유로는 2018년 11월 에마뉘엘 마크롱 대통령의 유류세 인상 발표로 인한   
‘노란조끼 시위’로 인해 내부 정치, 경제의 상태가 불안정 하다고 평가를 받는 프랑스의 경우도 기존 35%에서 24%로 신뢰도를 유지하고 있다. 

즉 시위가 일어나고 있는 프랑스, 또한 국가 경제가 위태로운 그리스 보다 더 낮은 언론 신뢰도를 가지고 있는 것이 한국이라는 뜻이다.

이토록 한국의 언론이 신뢰를 받지 못하는 이유를 살펴보기 위해,  
언론의 신뢰도가 낮은 이유에 대해서 일반 국민과 언론인들은 어떻게 생각하는지 조사한 보고서를 살펴본 결과는 다음과 같다. 



해당 보고서에서 일반국민과 언론인들이 언론 신뢰도가 낮은 이유로 평가를 한 것은 바로 특정 정치 세력에 편향된 보도태도이다. 

여기서 주목할 만한 점은 일반국민과 언론인이 비슷한 비율로 나온 다른 항목과는 다르게,  
특정 정치 세력에 편향된 보도태도 항목에서 일반국민보다 언론인이 무려 6.7%나 높게 나타났다는 사실이다. 

이는 해당 업무를 직접 하고 있는 언론인들이 간접적으로 겪고 있는 일반국민들 보다  
해당 원인을 더 정확하게 파악하는 혹은 이러한 사항으로 문제를 겪고 있다는 것으로 해석이 가능하다.

그 다음으로는 2018 에델만 신뢰도 지표 조사 항목 중 하나인 언론기관에 대한 회의감 조사에 대한 결과는 다음과 같다.



청중들이 생각하기에 언론사가 지나치게 집중해서 다루는 부분의 비율이 시청률을 위한 뉴스, 속보 항목에서 각각 70%, 67%로 나타난 가운데,   
그 다음 항목으로 바로 나타나는 것이 바로 정치 항목이다. 

특정 이념 지지에 지나치게 집중해서 다룬다고 생각한 인원이 응답자의 약 61%를 차지하고 있다는 것은,   
앞서 제시된 보고서의 한 항목과 일치하게 된다는 것을 발견할 수 있게 된다. 

즉 언론사가 특정 이념 대해 지나치게 집중해서 다루고 있고,   
이러한 언론사의 보도 태도로 인해 일반국민들은 언론사를 신뢰하지 못하게 되면서, 한국의 언론사가 신뢰를 받지 못한다고 평가가 가능하다.

특히 이러한 정치적인 이념 지지가 많이 드러나는 대표적인 언어가 바로 조중동, 한경오라는 단어이다. 

조중동은 조선일보, 중앙일보, 동아일보를 지칭하는 약자로 보수세력을 대표하는 언론사로 많이 굳혀져 있고,  
주로 진보세력이 보수언론을 비판하기 위한 표현으로 자주 사용이 되고 있다. 

이와 반대되는 한경오는 한겨레 신문, 경향 신문, 오마이뉴스를 지칭하는 약자로,  
보수 세력이 진보 언론을 비판하기 위한 표현이 아닌 진보 세력이 진보 언론을 비판하기 위한 표현으로 사용이 되었다.  

하지만 이러한 단어들은 모두 진보세력이 언론사를 비판하기 위해서 만든 개념으로,  
실제로 해당 언론사들이 보수적인 성향, 진보적인 성향을 띄는지를 확인하기에는 무리가 있으며,  
특히 해당 언론사의 각 기사들 마다 보수적, 진보적인 입장을 취한다고 판단을 내리기에도 비약적인 측면이 존재한다. 

그렇기 때문에, 이번 프로젝트를 통해서, 진보 정당을 대표하는 더불어민주당, 보수 정당을 대표하는 자유한국당의 논평 및 칼럼을 이용하여,  
글에서 보수, 진보 성향의 비율을 제시하는 분류 모델을 만들고 이를 시스템화 하는 과정을 시행하려고 한다. 

## 연구 방법

현재 분류를 하고 싶은 것은 사전적으로 정의가 된 보수, 진보가 아닌 현재 언론사가 가지고 있는 특정 정당에 대한 성향이기 때문에,  
보수 정당 중 가장 대표적인 정당인 자유한국당과 이와 반대되게 진보 정당 중 가장 대표적인 더불어민주당을 선정하게 되었다. 

그리고 각 정당에서 해당 정당의 입장 및 정치적인 성향을 가장 잘 나타내는 것은 논평 혹은 브리핑이라고 판단을 하게 되어,  
자유한국당 과 더불어민주당 의 논평 및 브리핑 게시물을 크롤링 하는 작업을 통해 데이터를 수집하였다. 

이후 각 게시물에서 정치적인 성향을 가질 만한 핵심 단어들을 전처리 및 명사 추출 작업을 통해 획득을 하는 작업을 수행하고,  
분류 모형인 로지스틱 회귀, 나이브 베이즈, svm 등의 기법을 활용하여, 비교 후 최적의 모형을 선택하고, 이를 시스템에 적용하여 서비스 하려고 한다.

Konlpy의 함수들 중에서 Okt (Open korea text), Hannanum, Kkma (Kind Korean Morpheme Analyzer), komoran 등이 있는 가운데,  
나중에 모델에 적용하기에 가장 좋은 자연어 처리 함수를 찾기 위해서, 샘플을 뽑아 분석을 해본 결과를 소개하려고 한다.

```python
con = sqlite3.connect(r"C:\Users\user\Desktop\project\libertykorea0813.db")
cur = con.cursor()
```

```python
cur.execute("""
    SELECT content
    FROM table1
    ORDER BY id ASC
""");

corpus = [_[0] for _ in cur.fetchall()]
```

그렇기 때문에 현재 정확한 판단 기준이 정해져 있지 않은데, 해당 UI를 제시하게 된다면,  
정치적 성향을 판단하는 기준이 없는 시점에 새로운 정치 성향 판단의 지표를 제시할 수 있을 것으로 보인다.

마지막으로, 새로운 정치 성향을 판단하는 지표로서도 사용이 가능하지만, 최종적인 판단은 기사를 읽는 사용자들의 몫이다.  
그렇기 때문에, 그러한 최종적인 판단을 돕고자, 정치성향을 분류해주는 것이 아닌 해당 정치성향으로 분류가 될 확률을 제시해준 것이다.  
제시된 확률을 바탕으로, 사용자는 해당 글이 어떠한 정치적 성향을 가지고 있는지 사전에 판단이 가능하기 때문에,  
해당 사건을 여러 개의 기사를 바탕으로 조사하고자 하는 경우, 혹은 알고자 하는 경우에,  
너무 편향적으로 제시가 된 기사들의 경우에는 미리 선별을 하고 읽을수 있도록 하는 보조적인 수단으로 사용이 가능하다.
